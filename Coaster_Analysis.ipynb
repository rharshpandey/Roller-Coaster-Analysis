```python
df = df[['coaster_name', 
    #'Length', 'Speed', 
         'Location', 'Status',
         #'Opening date',
      # 'Type', 
    'Manufacturer', 
    #'Height restriction', 'Model', 'Height',
      # 'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
      # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
      # 'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
      # 'Closing date', 
    #'Opened', 
    #'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 
         'latitude', 'longitude',
    'Type_Main',
       'opening_date_clean', 
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
    #'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df.rename(columns={'coaster_name':'Coaster_Name',
                  'year_introduced':'Year_Introduced',
                  'latitude':'Latitude',
                  'longitude':'Longitude',
                  'opening_date_clean':'Opening_Date',
                  'speed_mph':'Speed_mph',
                  'height_ft':'Height_ft',
                  'Inversion_clean':'Inversion',
                  'Gforce_clean':'Gforce'},inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df = df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True).copy()
```

```python
df['Year_Introduced'].value_counts()
```

```python
ax = df['Year_Introduced'].value_counts().head(10).plot(kind='bar',title = 'Top 10 Years Coaster Introduced')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Counts')
```

```python
df['Speed_mph'].plot(kind='hist',
                    bins=20,
                    title='Coaster Speed (mph)',
                    color='g')
```

```python
ax = df['Speed_mph'].plot(kind='kde',
                         title='Coaster Speed (mph)',
                         color='blue')
ax.set_xlabel('Speed (mph)')
```

```python
df.plot(kind='scatter',
       x='Speed_mph',
       y='Height_ft',
       title='Coaster Speed vs Height')
plt.show()
```

```python
sns.scatterplot(x='Speed_mph',
               y='Height_ft',
                hue='Year_Introduced',
               data=df)
plt.show()
```

```python
sns.pairplot(df,vars=['Year_Introduced','Speed_mph',
                     'Height_ft','Inversions_clean','Gforce'],
            hue='Type_Main')
plt.show()
```

```python
df_corr=df[['Year_Introduced','Speed_mph',
'Height_ft','Inversions_clean','Gforce']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr,annot=True)
```

```python
ax = df.query('Location != "Other"') \
        .groupby('Location')['Speed_mph'] \
        .agg(['mean','count']) \
        .query('count >= 10') \
        .sort_values('mean') \
        .plot(kind='barh',figsize=(12,5),title='Average Speed of Coast by Location')
ax.set_xlabel("Average Speed of Coaster")
plt.show()
```
